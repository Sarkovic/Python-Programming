```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gapminder = pd.read_csv("Datasets\\gapminder.csv")

gapminder
```

```python
# Create a scatter plot with GDP per capita on the x-axis and life expectancy on the y-axis
plt.scatter(gapminder["gdp_cap"], gapminder["life_exp"])

# Set the scale of the x-axis to logarithmic
plt.xscale('log')

# Label the x-axis and y-axis
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")

# Set a title for the plot
plt.title("World Development in 2007")

# Display the plot
plt.show()
```

```python
# Create a scatter plot with GDP per capita on the x-axis and life expectancy on the y-axis
plt.scatter(gapminder["gdp_cap"], gapminder["life_exp"])

# Set the scale of the x-axis to logarithmic
plt.xscale('log')

# Label the x-axis and y-axis
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")

# Set a title for the plot
plt.title("World Development in 2007")

# Define the ticks
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# Display the plot
plt.show()
```

```python
# Import numpy as np
import numpy as np

# Store pop as a numpy array: np_pop
np_pop = np.array(gapminder["population"])

# Double np_pop
np_pop = (np_pop * 2) / 1000000

# Update: set s argument to np_pop
plt.scatter(gapminder["gdp_cap"], gapminder["life_exp"], s=np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show()
```

```python
# Define a dictionary mapping continents to colors
continent_colors = {
    'Asia': 'red',
    'Europe': 'green',
    'Africa': 'blue',
    'Americas': 'yellow',
    'Oceania': 'black'
}

col = gapminder['cont'].map(continent_colors)

# Store pop as a numpy array: np_pop
np_pop = np.array(gapminder["population"])

# Double np_pop
np_pop = (np_pop * 2) / 1000000

# Update: set s argument to np_pop
plt.scatter(gapminder["gdp_cap"], gapminder["life_exp"], s=np_pop, c=col, alpha=0.8)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show()
```
