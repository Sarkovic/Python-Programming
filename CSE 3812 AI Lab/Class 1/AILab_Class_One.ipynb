# String

```python
message = "Python Programming"

# String length
print(len(message))

# String char print
print(message[1])

# Negative index means reverse travelling
print(message[-1])

# String slicing print
print(message[0:3])
# First or Last index is not mandatory
print(message[:5])

print(message[5:])

# String methods
print(message.upper())
print(message.capitalize())

print("Index =", message.find("Pro"))

print(message.replace("P", "J"))
```

# Operator
1. Arithmatic
2. Logical
3. Relational
4. Shorthand

```python
# Arithmatic Operator
x = 11
y = 5
print(x + y)

print(x - y)

print(x * y)

print(x / y)

print(x % y)

print(x ** y)

print(x // y)
```

```python
x = int(input("Enter a number:"))

# Taking last digit
first = x % 10
# Storing it in a variable
sum = first * 100
# Removing last digit by integer divide
x //= 10
# Same process as above
second = x % 10
sum += second * 10
# After this only one digit will remain, the first digit
x //= 10 
# Storing it in the sum variable
sum += x

print(sum)
```

```python
# Relational Operator: >, <, >=, <=, ==, !=

# Logical Operator: and, or, not
print(2 < 5 and 2 == 5)

print(2 < 5 and not 2 == 5)
```

```python
# Conditional Operator

gender = 'F'
types = "male"

if gender == 'F' or gender == 'f':
  print("Welcome to our beauty parlor!")
elif gender == 'm' and types == "male":
  print("Okay")
else:
  print("Sorry! you are not allowed")

# In operation
student = "Rahad Khan"
if 'A' in student:
  print("Found!")
else:
  print("Not Found")
```

# Loop

```python
# range(start, end, stepsize)
for i in range(1, 10, 2):
  print(f"i: {i}")

print("\n")
# Print start pattern
for i in range(1, 6, 1):
  print("*"*i)
```

```python
# Print Pyramid Star Pattern
for i in range(1, 5, 1):
  print(" "*(5 - i), "* "*i)
```
